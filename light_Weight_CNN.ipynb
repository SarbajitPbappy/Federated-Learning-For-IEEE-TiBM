```python
import os
from pathlib import Path
import itertools
import random
import json
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, roc_curve, auc

# ==========================================================
# 1ï¸âƒ£ Paths
# ==========================================================
DATASET_DIR = Path('/Volumes/Sarbajit/Model Optimization/Data/Augmented/Breast Cancer')
RESULTS_ROOT = Path.cwd() / "CustomCNN"
RESULTS_ROOT.mkdir(parents=True, exist_ok=True)

# ==========================================================
# 2ï¸âƒ£ Build DataFrame & Split
# ==========================================================
def build_image_dataframe(dataset_dir: Path):
    filepaths = list(dataset_dir.glob('**/*.jpg'))
    labels = [os.path.split(os.path.split(f)[0])[1] for f in filepaths]
    df = pd.DataFrame({'Filepath': [str(f) for f in filepaths], 'Label': labels})
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    return df

image_df = build_image_dataframe(DATASET_DIR)
print("Total images:", len(image_df))
print("Classes:", sorted(image_df['Label'].unique()))

# Stratified 70/15/15 split
train_val_df, test_df = train_test_split(image_df, train_size=0.85, random_state=42, stratify=image_df['Label'])
train_df, val_df = train_test_split(train_val_df, train_size=0.8235, random_state=42, stratify=train_val_df['Label'])
print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# ==========================================================
# 3ï¸âƒ£ Data Generators
# ==========================================================
def create_generators(train_df, val_df, test_df, config):
    num_classes = train_df['Label'].nunique()
    class_mode = 'binary' if num_classes == 2 else 'categorical'

    preprocess = tf.keras.applications.mobilenet_v2.preprocess_input

    train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        preprocessing_function=preprocess,
        rotation_range=config.get('rotation_range', 0),
        zoom_range=config.get('zoom_range', 0.0),
        horizontal_flip=config.get('horizontal_flip', False),
        brightness_range=config.get('brightness_range', None)
    )

    val_test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=preprocess)

    target_size = config['input_shape'][:2]
    batch_size = config.get('batch_size', 32)

    train_gen = train_datagen.flow_from_dataframe(
        train_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=True
    )

    val_gen = val_test_datagen.flow_from_dataframe(
        val_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    test_gen = val_test_datagen.flow_from_dataframe(
        test_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    return train_gen, val_gen, test_gen, num_classes

# ==========================================================
# ðŸ”¹ Attention Blocks (Keras-friendly)
# ==========================================================

def se_block(x, ratio=16):
    filters = x.shape[-1]
    se = layers.GlobalAveragePooling2D()(x)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, 1, filters))(se)
    return layers.Multiply()([x, se])

def cbam_block(x, ratio=8):
    channel = x.shape[-1]

    # Channel attention
    avg_pool = layers.GlobalAveragePooling2D()(x)
    max_pool = layers.GlobalMaxPooling2D()(x)
    shared_dense_one = layers.Dense(channel // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
    shared_dense_two = layers.Dense(channel, kernel_initializer='he_normal', use_bias=False)
    avg_out = shared_dense_two(shared_dense_one(avg_pool))
    max_out = shared_dense_two(shared_dense_one(max_pool))
    channel_attention = layers.Add()([avg_out, max_out])
    channel_attention = layers.Activation('sigmoid')(channel_attention)
    channel_attention = layers.Reshape((1, 1, channel))(channel_attention)
    x_ca = layers.Multiply()([x, channel_attention])

    # Spatial attention
    avg_pool_spatial = layers.Lambda(lambda t: tf.reduce_mean(t, axis=-1, keepdims=True))(x_ca)
    max_pool_spatial = layers.Lambda(lambda t: tf.reduce_max(t, axis=-1, keepdims=True))(x_ca)
    concat = layers.Concatenate(axis=-1)([avg_pool_spatial, max_pool_spatial])
    spatial_attention = layers.Conv2D(1, kernel_size=7, padding='same', activation='sigmoid')(concat)
    return layers.Multiply()([x_ca, spatial_attention])

def self_attention_block(x):
    """
    Safe self-attention using Lambda to wrap raw TF ops.
    Only use when spatial size is small (e.g., <= 14x14).
    """
    channels = x.shape[-1]
    f = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    g = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    h = layers.Conv2D(channels, 1, padding='same')(x)

    s = layers.Lambda(lambda tensors: tf.matmul(
        tf.reshape(tensors[0], [tf.shape(tensors[0])[0], -1, tf.shape(tensors[0])[-1]]),
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]]),
        transpose_b=True
    ))([f, g])

    beta = layers.Activation('softmax')(s)

    o = layers.Lambda(lambda tensors: tf.matmul(
        tensors[0],
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]])
    ))([beta, h])

    o = layers.Lambda(lambda tensors: tf.reshape(tensors[0], tf.shape(tensors[1])))([o, x])
    return layers.Add()([o, x])

# ==========================================================
# ðŸ”¹ Final Custom CNN Model â€” MEMORY-SAFE VERSION
# ==========================================================

def build_final_custom_cnn(input_shape=(224, 224, 3), num_classes=2):
    inputs = layers.Input(shape=input_shape)
    x = inputs

    filters = [32, 64, 128, 256]
    kernel_sizes = [3, 3, 3, 3]

    for i, (f, k) in enumerate(zip(filters, kernel_sizes)):
        x = layers.Conv2D(f, k, padding='same', kernel_initializer='he_normal',
                          kernel_regularizer=regularizers.l2(1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)

        # âœ… Apply SE in all blocks
        x = se_block(x, ratio=8)

        # âœ… Apply CBAM only in deeper blocks (i >= 2 â†’ 56x56 and smaller)
        if i >= 2:
            x = cbam_block(x, ratio=8)

        # âœ… Apply self-attention ONLY in the last block (14x14)
        if i == 3:
            x = self_attention_block(x)

        x = layers.MaxPooling2D(pool_size=2)(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)

    if num_classes == 2:
        outputs = layers.Dense(1, activation='sigmoid')(x)
        loss_fn = 'binary_crossentropy'
        metrics = ['accuracy', tf.keras.metrics.AUC(name='auc')]
    else:
        outputs = layers.Dense(num_classes, activation='softmax')(x)
        loss_fn = 'categorical_crossentropy'
        metrics = ['accuracy']

    model = models.Model(inputs, outputs)
    opt = optimizers.Adam(learning_rate=0.0001)
    model.compile(optimizer=opt, loss=loss_fn, metrics=metrics)
    return model

# ==========================================================
# 4ï¸âƒ£ Training Configuration
# ==========================================================
config = {
    'input_shape': (224, 224, 3),
    'batch_size': 16,  # Reduced to be safe (was 32)
    'rotation_range': 15,
    'zoom_range': 0.1,
    'horizontal_flip': True,
    'brightness_range': [0.8, 1.2]
}

train_gen, val_gen, test_gen, num_classes = create_generators(train_df, val_df, test_df, config)

model = build_final_custom_cnn(input_shape=config['input_shape'], num_classes=num_classes)
model.summary()

# ==========================================================
# 5ï¸âƒ£ Model Training
# ==========================================================
EPOCHS = 30

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    verbose=1
)

# ==========================================================
# 6ï¸âƒ£ Plot Training Curves
# ==========================================================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig(RESULTS_ROOT / "training_curves.png")
plt.show()

# ==========================================================
# 7ï¸âƒ£ Evaluation on Test Set
# ==========================================================
test_steps = len(test_gen)
test_preds = model.predict(test_gen, steps=test_steps, verbose=1)
true_labels = test_gen.classes
class_indices = test_gen.class_indices
class_names = list(class_indices.keys())

if num_classes == 2:
    y_pred_prob = test_preds.ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)
else:
    y_pred_prob = test_preds
    y_pred = np.argmax(test_preds, axis=1)

# Trim predictions if generator repeated last batch
if len(y_pred) > len(true_labels):
    y_pred = y_pred[:len(true_labels)]
    y_pred_prob = y_pred_prob[:len(true_labels)]

# ==========================================================
# 8ï¸âƒ£ Confusion Matrix & Classification Report
# ==========================================================
cm = confusion_matrix(true_labels, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig(RESULTS_ROOT / "confusion_matrix.png")
plt.show()

print("ðŸ“‹ Classification Report:")
print(classification_report(true_labels, y_pred, target_names=class_names))

# ==========================================================
# 9ï¸âƒ£ ROC Curve
# ==========================================================
plt.figure(figsize=(8, 6))
if num_classes == 2:
    fpr, tpr, _ = roc_curve(true_labels, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
else:
    y_true_bin = tf.keras.utils.to_categorical(true_labels, num_classes=num_classes)
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC={roc_auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig(RESULTS_ROOT / "roc_curve.png")
plt.show()
```

```python
import os
from pathlib import Path
import itertools
import random
import json
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, roc_curve, auc

# ==========================================================
# 1ï¸âƒ£ Paths
# ==========================================================
DATASET_DIR = Path('/Volumes/Sarbajit/Model Optimization/Data/Augmented/Breast Cancer')
RESULTS_ROOT = Path.cwd() / "CustomCNN2"
RESULTS_ROOT.mkdir(parents=True, exist_ok=True)

# ==========================================================
# 2ï¸âƒ£ Build DataFrame & Split
# ==========================================================
def build_image_dataframe(dataset_dir: Path):
    filepaths = list(dataset_dir.glob('**/*.jpg'))
    labels = [os.path.split(os.path.split(f)[0])[1] for f in filepaths]
    df = pd.DataFrame({'Filepath': [str(f) for f in filepaths], 'Label': labels})
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    return df

image_df = build_image_dataframe(DATASET_DIR)
print("Total images:", len(image_df))
print("Classes:", sorted(image_df['Label'].unique()))

# Stratified 70/15/15 split
train_val_df, test_df = train_test_split(image_df, train_size=0.85, random_state=42, stratify=image_df['Label'])
train_df, val_df = train_test_split(train_val_df, train_size=0.8235, random_state=42, stratify=train_val_df['Label'])
print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# ==========================================================
# 3ï¸âƒ£ Data Generators
# ==========================================================
def create_generators(train_df, val_df, test_df, config):
    num_classes = train_df['Label'].nunique()
    class_mode = 'binary' if num_classes == 2 else 'categorical'

    preprocess = tf.keras.applications.mobilenet_v2.preprocess_input

    train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        preprocessing_function=preprocess,
        rotation_range=config.get('rotation_range', 0),
        zoom_range=config.get('zoom_range', 0.0),
        horizontal_flip=config.get('horizontal_flip', False),
        brightness_range=config.get('brightness_range', None)
    )

    val_test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=preprocess)

    target_size = config['input_shape'][:2]
    batch_size = config.get('batch_size', 32)

    train_gen = train_datagen.flow_from_dataframe(
        train_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=True
    )

    val_gen = val_test_datagen.flow_from_dataframe(
        val_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    test_gen = val_test_datagen.flow_from_dataframe(
        test_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    return train_gen, val_gen, test_gen, num_classes

# ==========================================================
# ðŸ”¹ Attention Blocks (Keras-friendly)
# ==========================================================

def se_block(x, ratio=16):
    filters = x.shape[-1]
    se = layers.GlobalAveragePooling2D()(x)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, 1, filters))(se)
    return layers.Multiply()([x, se])

def cbam_block(x, ratio=8):
    channel = x.shape[-1]

    # Channel attention
    avg_pool = layers.GlobalAveragePooling2D()(x)
    max_pool = layers.GlobalMaxPooling2D()(x)
    shared_dense_one = layers.Dense(channel // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
    shared_dense_two = layers.Dense(channel, kernel_initializer='he_normal', use_bias=False)
    avg_out = shared_dense_two(shared_dense_one(avg_pool))
    max_out = shared_dense_two(shared_dense_one(max_pool))
    channel_attention = layers.Add()([avg_out, max_out])
    channel_attention = layers.Activation('sigmoid')(channel_attention)
    channel_attention = layers.Reshape((1, 1, channel))(channel_attention)
    x_ca = layers.Multiply()([x, channel_attention])

    # Spatial attention
    avg_pool_spatial = layers.Lambda(lambda t: tf.reduce_mean(t, axis=-1, keepdims=True))(x_ca)
    max_pool_spatial = layers.Lambda(lambda t: tf.reduce_max(t, axis=-1, keepdims=True))(x_ca)
    concat = layers.Concatenate(axis=-1)([avg_pool_spatial, max_pool_spatial])
    spatial_attention = layers.Conv2D(1, kernel_size=7, padding='same', activation='sigmoid')(concat)
    return layers.Multiply()([x_ca, spatial_attention])

def self_attention_block(x):
    """
    Safe self-attention using Lambda to wrap raw TF ops.
    Only use when spatial size is small (e.g., <= 14x14).
    """
    channels = x.shape[-1]
    f = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    g = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    h = layers.Conv2D(channels, 1, padding='same')(x)

    s = layers.Lambda(lambda tensors: tf.matmul(
        tf.reshape(tensors[0], [tf.shape(tensors[0])[0], -1, tf.shape(tensors[0])[-1]]),
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]]),
        transpose_b=True
    ))([f, g])

    beta = layers.Activation('softmax')(s)

    o = layers.Lambda(lambda tensors: tf.matmul(
        tensors[0],
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]])
    ))([beta, h])

    o = layers.Lambda(lambda tensors: tf.reshape(tensors[0], tf.shape(tensors[1])))([o, x])
    return layers.Add()([o, x])

# ==========================================================
# ðŸ”¹ Final Custom CNN Model â€” MEMORY-SAFE VERSION
# ==========================================================

def build_final_custom_cnn(input_shape=(224, 224, 3), num_classes=2):
    inputs = layers.Input(shape=input_shape)
    x = inputs

    filters = [32, 64, 128, 256]
    kernel_sizes = [3, 3, 3, 3]

    for i, (f, k) in enumerate(zip(filters, kernel_sizes)):
        x = layers.Conv2D(f, k, padding='same', kernel_initializer='he_normal',
                          kernel_regularizer=regularizers.l2(1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)

        # âœ… Apply SE in all blocks
        x = se_block(x, ratio=8)

        # âœ… Apply CBAM only in deeper blocks (i >= 2 â†’ 56x56 and smaller)
        if i >= 2:
            x = cbam_block(x, ratio=8)

        # âœ… Apply self-attention ONLY in the last block (14x14)
        if i == 3:
            x = self_attention_block(x)

        x = layers.MaxPooling2D(pool_size=2)(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)

    if num_classes == 2:
        outputs = layers.Dense(1, activation='sigmoid')(x)
        loss_fn = tf.keras.losses.BinaryCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy', tf.keras.metrics.AUC(name='auc')]
    else:
        outputs = layers.Dense(num_classes, activation='softmax')(x)
        loss_fn = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy']

    model = models.Model(inputs, outputs)
    opt = optimizers.Adam(learning_rate=0.0001)
    model.compile(optimizer=opt, loss=loss_fn, metrics=metrics)
    return model

# ==========================================================
# 4ï¸âƒ£ Training Configuration
# ==========================================================
config = {
    'input_shape': (224, 224, 3),
    'batch_size': 16,  # Reduced to be safe (was 32)
    'rotation_range': 15,
    'zoom_range': 0.1,
    'horizontal_flip': True,
    'brightness_range': [0.8, 1.2]
}
# Add callbacks
callbacks = [
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    ),
    EarlyStopping(
        monitor='val_loss',
        patience=6,
        restore_best_weights=True,
        verbose=1
    )
]
train_gen, val_gen, test_gen, num_classes = create_generators(train_df, val_df, test_df, config)

model = build_final_custom_cnn(input_shape=config['input_shape'], num_classes=num_classes)
# model.summary()

# ==========================================================
# 5ï¸âƒ£ Model Training
# ==========================================================
EPOCHS = 30

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=30,
    callbacks=callbacks,  
    verbose=1
)

# ==========================================================
# 6ï¸âƒ£ Plot Training Curves
# ==========================================================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig(RESULTS_ROOT / "training_curves.png")
plt.show()

# ==========================================================
# 7ï¸âƒ£ Evaluation on Test Set
# ==========================================================
test_steps = len(test_gen)
test_preds = model.predict(test_gen, steps=test_steps, verbose=1)
true_labels = test_gen.classes
class_indices = test_gen.class_indices
class_names = list(class_indices.keys())

if num_classes == 2:
    y_pred_prob = test_preds.ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)
else:
    y_pred_prob = test_preds
    y_pred = np.argmax(test_preds, axis=1)

# Trim predictions if generator repeated last batch
if len(y_pred) > len(true_labels):
    y_pred = y_pred[:len(true_labels)]
    y_pred_prob = y_pred_prob[:len(true_labels)]

# ==========================================================
# 8ï¸âƒ£ Confusion Matrix & Classification Report
# ==========================================================
cm = confusion_matrix(true_labels, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig(RESULTS_ROOT / "confusion_matrix.png")
plt.show()

print("ðŸ“‹ Classification Report:")
print(classification_report(true_labels, y_pred, target_names=class_names))

# ==========================================================
# 9ï¸âƒ£ ROC Curve
# ==========================================================
plt.figure(figsize=(8, 6))
if num_classes == 2:
    fpr, tpr, _ = roc_curve(true_labels, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
else:
    y_true_bin = tf.keras.utils.to_categorical(true_labels, num_classes=num_classes)
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC={roc_auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig(RESULTS_ROOT / "roc_curve.png")
plt.show()
```

```python
import os
from pathlib import Path
import itertools
import random
import json
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, roc_curve, auc

# ==========================================================
# 1ï¸âƒ£ Paths
# ==========================================================
DATASET_DIR = Path('/Volumes/Sarbajit/Model Optimization/Data/Augmented/Kidney Cancer')
RESULTS_ROOT = Path.cwd() / "CustomCNN_kidney"
RESULTS_ROOT.mkdir(parents=True, exist_ok=True)

# ==========================================================
# 2ï¸âƒ£ Build DataFrame & Split
# ==========================================================
def build_image_dataframe(dataset_dir: Path):
    filepaths = list(dataset_dir.glob("**/*.jpg")) + list(dataset_dir.glob("**/*.png")) + list(dataset_dir.glob("**/*.jpeg"))
    labels = [os.path.split(os.path.split(f)[0])[1] for f in filepaths]
    df = pd.DataFrame({'Filepath': [str(f) for f in filepaths], 'Label': labels})
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    return df

image_df = build_image_dataframe(DATASET_DIR)
print("Total images:", len(image_df))
print("Classes:", sorted(image_df['Label'].unique()))

# Stratified 70/15/15 split
train_val_df, test_df = train_test_split(image_df, train_size=0.85, random_state=42, stratify=image_df['Label'])
train_df, val_df = train_test_split(train_val_df, train_size=0.8235, random_state=42, stratify=train_val_df['Label'])
print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# ==========================================================
# 3ï¸âƒ£ Data Generators
# ==========================================================
def create_generators(train_df, val_df, test_df, config):
    num_classes = train_df['Label'].nunique()
    class_mode = 'binary' if num_classes == 2 else 'categorical'

    preprocess = tf.keras.applications.mobilenet_v2.preprocess_input

    train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        preprocessing_function=preprocess,
        rotation_range=config.get('rotation_range', 0),
        zoom_range=config.get('zoom_range', 0.0),
        horizontal_flip=config.get('horizontal_flip', False),
        brightness_range=config.get('brightness_range', None)
    )

    val_test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=preprocess)

    target_size = config['input_shape'][:2]
    batch_size = config.get('batch_size', 32)

    train_gen = train_datagen.flow_from_dataframe(
        train_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=True
    )

    val_gen = val_test_datagen.flow_from_dataframe(
        val_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    test_gen = val_test_datagen.flow_from_dataframe(
        test_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    return train_gen, val_gen, test_gen, num_classes

# ==========================================================
# ðŸ”¹ Attention Blocks (Keras-friendly)
# ==========================================================

def se_block(x, ratio=16):
    filters = x.shape[-1]
    se = layers.GlobalAveragePooling2D()(x)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, 1, filters))(se)
    return layers.Multiply()([x, se])

def cbam_block(x, ratio=8):
    channel = x.shape[-1]

    # Channel attention
    avg_pool = layers.GlobalAveragePooling2D()(x)
    max_pool = layers.GlobalMaxPooling2D()(x)
    shared_dense_one = layers.Dense(channel // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
    shared_dense_two = layers.Dense(channel, kernel_initializer='he_normal', use_bias=False)
    avg_out = shared_dense_two(shared_dense_one(avg_pool))
    max_out = shared_dense_two(shared_dense_one(max_pool))
    channel_attention = layers.Add()([avg_out, max_out])
    channel_attention = layers.Activation('sigmoid')(channel_attention)
    channel_attention = layers.Reshape((1, 1, channel))(channel_attention)
    x_ca = layers.Multiply()([x, channel_attention])

    # Spatial attention
    avg_pool_spatial = layers.Lambda(lambda t: tf.reduce_mean(t, axis=-1, keepdims=True))(x_ca)
    max_pool_spatial = layers.Lambda(lambda t: tf.reduce_max(t, axis=-1, keepdims=True))(x_ca)
    concat = layers.Concatenate(axis=-1)([avg_pool_spatial, max_pool_spatial])
    spatial_attention = layers.Conv2D(1, kernel_size=7, padding='same', activation='sigmoid')(concat)
    return layers.Multiply()([x_ca, spatial_attention])

def self_attention_block(x):
    """
    Safe self-attention using Lambda to wrap raw TF ops.
    Only use when spatial size is small (e.g., <= 14x14).
    """
    channels = x.shape[-1]
    f = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    g = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    h = layers.Conv2D(channels, 1, padding='same')(x)

    s = layers.Lambda(lambda tensors: tf.matmul(
        tf.reshape(tensors[0], [tf.shape(tensors[0])[0], -1, tf.shape(tensors[0])[-1]]),
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]]),
        transpose_b=True
    ))([f, g])

    beta = layers.Activation('softmax')(s)

    o = layers.Lambda(lambda tensors: tf.matmul(
        tensors[0],
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]])
    ))([beta, h])

    o = layers.Lambda(lambda tensors: tf.reshape(tensors[0], tf.shape(tensors[1])))([o, x])
    return layers.Add()([o, x])

# ==========================================================
# ðŸ”¹ Final Custom CNN Model â€” MEMORY-SAFE VERSION
# ==========================================================

def build_final_custom_cnn(input_shape=(224, 224, 3), num_classes=4):
    inputs = layers.Input(shape=input_shape)
    x = inputs

    filters = [32, 64, 128, 256]
    kernel_sizes = [3, 3, 3, 3]

    for i, (f, k) in enumerate(zip(filters, kernel_sizes)):
        x = layers.Conv2D(f, k, padding='same', kernel_initializer='he_normal',
                          kernel_regularizer=regularizers.l2(1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)

        # âœ… Apply SE in all blocks
        x = se_block(x, ratio=8)

        # âœ… Apply CBAM only in deeper blocks (i >= 2 â†’ 56x56 and smaller)
        if i >= 2:
            x = cbam_block(x, ratio=8)

        # âœ… Apply self-attention ONLY in the last block (14x14)
        if i == 3:
            x = self_attention_block(x)

        x = layers.MaxPooling2D(pool_size=2)(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)

    if num_classes == 2:
        outputs = layers.Dense(1, activation='sigmoid')(x)
        loss_fn = tf.keras.losses.BinaryCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy', tf.keras.metrics.AUC(name='auc')]
    else:
        outputs = layers.Dense(num_classes, activation='softmax')(x)
        loss_fn = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy']

    model = models.Model(inputs, outputs)
    opt = optimizers.Adam(learning_rate=0.0001)
    model.compile(optimizer=opt, loss=loss_fn, metrics=metrics)
    return model

# ==========================================================
# 4ï¸âƒ£ Training Configuration
# ==========================================================
config = {
    'input_shape': (224, 224, 3),
    'batch_size': 16,  
    'rotation_range': 15,
    'zoom_range': 0.1,
    'horizontal_flip': True,
    'brightness_range': [0.8, 1.2]
}
# Add callbacks
callbacks = [
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    )
]
train_gen, val_gen, test_gen, num_classes = create_generators(train_df, val_df, test_df, config)

model = build_final_custom_cnn(input_shape=config['input_shape'], num_classes=num_classes)
# model.summary()

# ==========================================================
# 5ï¸âƒ£ Model Training
# ==========================================================
EPOCHS = 100

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks,  
    verbose=1
)

# ==========================================================
# 6ï¸âƒ£ Plot Training Curves
# ==========================================================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig(RESULTS_ROOT / "training_curves.png")
plt.show()

# ==========================================================
# 7ï¸âƒ£ Evaluation on Test Set
# ==========================================================
test_steps = len(test_gen)
test_preds = model.predict(test_gen, steps=test_steps, verbose=1)
true_labels = test_gen.classes
class_indices = test_gen.class_indices
class_names = list(class_indices.keys())

if num_classes == 2:
    y_pred_prob = test_preds.ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)
else:
    y_pred_prob = test_preds
    y_pred = np.argmax(test_preds, axis=1)

# Trim predictions if generator repeated last batch
if len(y_pred) > len(true_labels):
    y_pred = y_pred[:len(true_labels)]
    y_pred_prob = y_pred_prob[:len(true_labels)]

# ==========================================================
# 8ï¸âƒ£ Confusion Matrix & Classification Report
# ==========================================================
cm = confusion_matrix(true_labels, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig(RESULTS_ROOT / "confusion_matrix.png")
plt.show()

print("ðŸ“‹ Classification Report:")
print(classification_report(true_labels, y_pred, target_names=class_names))

# ==========================================================
# 9ï¸âƒ£ ROC Curve
# ==========================================================
plt.figure(figsize=(8, 6))
if num_classes == 2:
    fpr, tpr, _ = roc_curve(true_labels, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
else:
    y_true_bin = tf.keras.utils.to_categorical(true_labels, num_classes=num_classes)
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC={roc_auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig(RESULTS_ROOT / "roc_curve.png")
plt.show()
model.save("Kidney_cancer_model.h5")
```

```python
import os
from pathlib import Path
import itertools
import random
import json
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, roc_curve, auc

# ==========================================================
# 1ï¸âƒ£ Paths
# ==========================================================
DATASET_DIR = Path('/Volumes/Sarbajit/Model Optimization/Data/Augmented/Lung and Colon Cancer')
RESULTS_ROOT = Path.cwd() / "CustomCNN_Lung"
RESULTS_ROOT.mkdir(parents=True, exist_ok=True)

# ==========================================================
# 2ï¸âƒ£ Build DataFrame & Split
# ==========================================================
def build_image_dataframe(dataset_dir: Path):
    filepaths = list(dataset_dir.glob("**/*.jpg")) + list(dataset_dir.glob("**/*.png")) + list(dataset_dir.glob("**/*.jpeg"))
    labels = [os.path.split(os.path.split(f)[0])[1] for f in filepaths]
    df = pd.DataFrame({'Filepath': [str(f) for f in filepaths], 'Label': labels})
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    return df

image_df = build_image_dataframe(DATASET_DIR)
print("Total images:", len(image_df))
print("Classes:", sorted(image_df['Label'].unique()))

# Stratified 70/15/15 split
train_val_df, test_df = train_test_split(image_df, train_size=0.85, random_state=42, stratify=image_df['Label'])
train_df, val_df = train_test_split(train_val_df, train_size=0.8235, random_state=42, stratify=train_val_df['Label'])
print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# ==========================================================
# 3ï¸âƒ£ Data Generators
# ==========================================================
def create_generators(train_df, val_df, test_df, config):
    num_classes = train_df['Label'].nunique()
    class_mode = 'binary' if num_classes == 2 else 'categorical'

    preprocess = tf.keras.applications.mobilenet_v2.preprocess_input

    train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        preprocessing_function=preprocess,
        rotation_range=config.get('rotation_range', 0),
        zoom_range=config.get('zoom_range', 0.0),
        horizontal_flip=config.get('horizontal_flip', False),
        brightness_range=config.get('brightness_range', None)
    )

    val_test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=preprocess)

    target_size = config['input_shape'][:2]
    batch_size = config.get('batch_size', 32)

    train_gen = train_datagen.flow_from_dataframe(
        train_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=True
    )

    val_gen = val_test_datagen.flow_from_dataframe(
        val_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    test_gen = val_test_datagen.flow_from_dataframe(
        test_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    return train_gen, val_gen, test_gen, num_classes

# ==========================================================
# ðŸ”¹ Attention Blocks (Keras-friendly)
# ==========================================================

def se_block(x, ratio=16):
    filters = x.shape[-1]
    se = layers.GlobalAveragePooling2D()(x)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, 1, filters))(se)
    return layers.Multiply()([x, se])

def cbam_block(x, ratio=8):
    channel = x.shape[-1]

    # Channel attention
    avg_pool = layers.GlobalAveragePooling2D()(x)
    max_pool = layers.GlobalMaxPooling2D()(x)
    shared_dense_one = layers.Dense(channel // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
    shared_dense_two = layers.Dense(channel, kernel_initializer='he_normal', use_bias=False)
    avg_out = shared_dense_two(shared_dense_one(avg_pool))
    max_out = shared_dense_two(shared_dense_one(max_pool))
    channel_attention = layers.Add()([avg_out, max_out])
    channel_attention = layers.Activation('sigmoid')(channel_attention)
    channel_attention = layers.Reshape((1, 1, channel))(channel_attention)
    x_ca = layers.Multiply()([x, channel_attention])

    # Spatial attention
    avg_pool_spatial = layers.Lambda(lambda t: tf.reduce_mean(t, axis=-1, keepdims=True))(x_ca)
    max_pool_spatial = layers.Lambda(lambda t: tf.reduce_max(t, axis=-1, keepdims=True))(x_ca)
    concat = layers.Concatenate(axis=-1)([avg_pool_spatial, max_pool_spatial])
    spatial_attention = layers.Conv2D(1, kernel_size=7, padding='same', activation='sigmoid')(concat)
    return layers.Multiply()([x_ca, spatial_attention])

def self_attention_block(x):
    """
    Safe self-attention using Lambda to wrap raw TF ops.
    Only use when spatial size is small (e.g., <= 14x14).
    """
    channels = x.shape[-1]
    f = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    g = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    h = layers.Conv2D(channels, 1, padding='same')(x)

    s = layers.Lambda(lambda tensors: tf.matmul(
        tf.reshape(tensors[0], [tf.shape(tensors[0])[0], -1, tf.shape(tensors[0])[-1]]),
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]]),
        transpose_b=True
    ))([f, g])

    beta = layers.Activation('softmax')(s)

    o = layers.Lambda(lambda tensors: tf.matmul(
        tensors[0],
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]])
    ))([beta, h])

    o = layers.Lambda(lambda tensors: tf.reshape(tensors[0], tf.shape(tensors[1])))([o, x])
    return layers.Add()([o, x])

# ==========================================================
# ðŸ”¹ Final Custom CNN Model â€” MEMORY-SAFE VERSION
# ==========================================================

def build_final_custom_cnn(input_shape=(224, 224, 3), num_classes=4):
    inputs = layers.Input(shape=input_shape)
    x = inputs

    filters = [32, 64, 128, 256]
    kernel_sizes = [3, 3, 3, 3]

    for i, (f, k) in enumerate(zip(filters, kernel_sizes)):
        x = layers.Conv2D(f, k, padding='same', kernel_initializer='he_normal',
                          kernel_regularizer=regularizers.l2(1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)

        # âœ… Apply SE in all blocks
        x = se_block(x, ratio=8)

        # âœ… Apply CBAM only in deeper blocks (i >= 2 â†’ 56x56 and smaller)
        if i >= 2:
            x = cbam_block(x, ratio=8)

        # âœ… Apply self-attention ONLY in the last block (14x14)
        if i == 3:
            x = self_attention_block(x)

        x = layers.MaxPooling2D(pool_size=2)(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)

    if num_classes == 2:
        outputs = layers.Dense(1, activation='sigmoid')(x)
        loss_fn = tf.keras.losses.BinaryCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy', tf.keras.metrics.AUC(name='auc')]
    else:
        outputs = layers.Dense(num_classes, activation='softmax')(x)
        loss_fn = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy']

    model = models.Model(inputs, outputs)
    opt = optimizers.Adam(learning_rate=0.0001)
    model.compile(optimizer=opt, loss=loss_fn, metrics=metrics)
    return model

# ==========================================================
# 4ï¸âƒ£ Training Configuration
# ==========================================================
config = {
    'input_shape': (224, 224, 3),
    'batch_size': 16,  
    'rotation_range': 15,
    'zoom_range': 0.1,
    'horizontal_flip': True,
    'brightness_range': [0.8, 1.2]
}
# Add callbacks
callbacks = [
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    )
]
train_gen, val_gen, test_gen, num_classes = create_generators(train_df, val_df, test_df, config)

model = build_final_custom_cnn(input_shape=config['input_shape'], num_classes=num_classes)
# model.summary()

# ==========================================================
# 5ï¸âƒ£ Model Training
# ==========================================================
EPOCHS = 100

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks,  
    verbose=1
)

# ==========================================================
# 6ï¸âƒ£ Plot Training Curves
# ==========================================================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig(RESULTS_ROOT / "training_curves.png")
plt.show()

# ==========================================================
# 7ï¸âƒ£ Evaluation on Test Set
# ==========================================================
test_steps = len(test_gen)
test_preds = model.predict(test_gen, steps=test_steps, verbose=1)
true_labels = test_gen.classes
class_indices = test_gen.class_indices
class_names = list(class_indices.keys())

if num_classes == 2:
    y_pred_prob = test_preds.ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)
else:
    y_pred_prob = test_preds
    y_pred = np.argmax(test_preds, axis=1)

# Trim predictions if generator repeated last batch
if len(y_pred) > len(true_labels):
    y_pred = y_pred[:len(true_labels)]
    y_pred_prob = y_pred_prob[:len(true_labels)]

# ==========================================================
# 8ï¸âƒ£ Confusion Matrix & Classification Report
# ==========================================================
cm = confusion_matrix(true_labels, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig(RESULTS_ROOT / "confusion_matrix.png")
plt.show()

print("ðŸ“‹ Classification Report:")
print(classification_report(true_labels, y_pred, target_names=class_names))

# ==========================================================
# 9ï¸âƒ£ ROC Curve
# ==========================================================
plt.figure(figsize=(8, 6))
if num_classes == 2:
    fpr, tpr, _ = roc_curve(true_labels, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
else:
    y_true_bin = tf.keras.utils.to_categorical(true_labels, num_classes=num_classes)
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC={roc_auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig(RESULTS_ROOT / "roc_curve.png")
plt.show()
model.save("lung_cancer_model.h5")
```

```python
import os
from pathlib import Path
import itertools
import random
import json
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, roc_curve, auc

# ==========================================================
# 1ï¸âƒ£ Paths
# ==========================================================
DATASET_DIR = Path('/Volumes/Sarbajit/Model Optimization/Data/Augmented/Oral Cancer')
RESULTS_ROOT = Path.cwd() / "CustomCNN_oral"
RESULTS_ROOT.mkdir(parents=True, exist_ok=True)

# ==========================================================
# 2ï¸âƒ£ Build DataFrame & Split
# ==========================================================
def build_image_dataframe(dataset_dir: Path):
    filepaths = list(dataset_dir.glob("**/*.jpg")) + list(dataset_dir.glob("**/*.png")) + list(dataset_dir.glob("**/*.jpeg"))
    labels = [os.path.split(os.path.split(f)[0])[1] for f in filepaths]
    df = pd.DataFrame({'Filepath': [str(f) for f in filepaths], 'Label': labels})
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    return df

image_df = build_image_dataframe(DATASET_DIR)
print("Total images:", len(image_df))
print("Classes:", sorted(image_df['Label'].unique()))

# Stratified 70/15/15 split
train_val_df, test_df = train_test_split(image_df, train_size=0.85, random_state=42, stratify=image_df['Label'])
train_df, val_df = train_test_split(train_val_df, train_size=0.8235, random_state=42, stratify=train_val_df['Label'])
print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# ==========================================================
# 3ï¸âƒ£ Data Generators
# ==========================================================
def create_generators(train_df, val_df, test_df, config):
    num_classes = train_df['Label'].nunique()
    class_mode = 'binary' if num_classes == 2 else 'categorical'

    preprocess = tf.keras.applications.mobilenet_v2.preprocess_input

    train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        preprocessing_function=preprocess,
        rotation_range=config.get('rotation_range', 0),
        zoom_range=config.get('zoom_range', 0.0),
        horizontal_flip=config.get('horizontal_flip', False),
        brightness_range=config.get('brightness_range', None)
    )

    val_test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=preprocess)

    target_size = config['input_shape'][:2]
    batch_size = config.get('batch_size', 32)

    train_gen = train_datagen.flow_from_dataframe(
        train_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=True
    )

    val_gen = val_test_datagen.flow_from_dataframe(
        val_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    test_gen = val_test_datagen.flow_from_dataframe(
        test_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    return train_gen, val_gen, test_gen, num_classes

# ==========================================================
# ðŸ”¹ Attention Blocks (Keras-friendly)
# ==========================================================

def se_block(x, ratio=16):
    filters = x.shape[-1]
    se = layers.GlobalAveragePooling2D()(x)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, 1, filters))(se)
    return layers.Multiply()([x, se])

def cbam_block(x, ratio=8):
    channel = x.shape[-1]

    # Channel attention
    avg_pool = layers.GlobalAveragePooling2D()(x)
    max_pool = layers.GlobalMaxPooling2D()(x)
    shared_dense_one = layers.Dense(channel // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
    shared_dense_two = layers.Dense(channel, kernel_initializer='he_normal', use_bias=False)
    avg_out = shared_dense_two(shared_dense_one(avg_pool))
    max_out = shared_dense_two(shared_dense_one(max_pool))
    channel_attention = layers.Add()([avg_out, max_out])
    channel_attention = layers.Activation('sigmoid')(channel_attention)
    channel_attention = layers.Reshape((1, 1, channel))(channel_attention)
    x_ca = layers.Multiply()([x, channel_attention])

    # Spatial attention
    avg_pool_spatial = layers.Lambda(lambda t: tf.reduce_mean(t, axis=-1, keepdims=True))(x_ca)
    max_pool_spatial = layers.Lambda(lambda t: tf.reduce_max(t, axis=-1, keepdims=True))(x_ca)
    concat = layers.Concatenate(axis=-1)([avg_pool_spatial, max_pool_spatial])
    spatial_attention = layers.Conv2D(1, kernel_size=7, padding='same', activation='sigmoid')(concat)
    return layers.Multiply()([x_ca, spatial_attention])

def self_attention_block(x):
    """
    Safe self-attention using Lambda to wrap raw TF ops.
    Only use when spatial size is small (e.g., <= 14x14).
    """
    channels = x.shape[-1]
    f = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    g = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    h = layers.Conv2D(channels, 1, padding='same')(x)

    s = layers.Lambda(lambda tensors: tf.matmul(
        tf.reshape(tensors[0], [tf.shape(tensors[0])[0], -1, tf.shape(tensors[0])[-1]]),
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]]),
        transpose_b=True
    ))([f, g])

    beta = layers.Activation('softmax')(s)

    o = layers.Lambda(lambda tensors: tf.matmul(
        tensors[0],
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]])
    ))([beta, h])

    o = layers.Lambda(lambda tensors: tf.reshape(tensors[0], tf.shape(tensors[1])))([o, x])
    return layers.Add()([o, x])

# ==========================================================
# ðŸ”¹ Final Custom CNN Model â€” MEMORY-SAFE VERSION
# ==========================================================

def build_final_custom_cnn(input_shape=(224, 224, 3), num_classes=2):
    inputs = layers.Input(shape=input_shape)
    x = inputs

    filters = [32, 64, 128, 256]
    kernel_sizes = [3, 3, 3, 3]

    for i, (f, k) in enumerate(zip(filters, kernel_sizes)):
        x = layers.Conv2D(f, k, padding='same', kernel_initializer='he_normal',
                          kernel_regularizer=regularizers.l2(1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)

        # âœ… Apply SE in all blocks
        x = se_block(x, ratio=8)

        # âœ… Apply CBAM only in deeper blocks (i >= 2 â†’ 56x56 and smaller)
        if i >= 2:
            x = cbam_block(x, ratio=8)

        # âœ… Apply self-attention ONLY in the last block (14x14)
        if i == 3:
            x = self_attention_block(x)

        x = layers.MaxPooling2D(pool_size=2)(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)

    if num_classes == 2:
        outputs = layers.Dense(1, activation='sigmoid')(x)
        loss_fn = tf.keras.losses.BinaryCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy', tf.keras.metrics.AUC(name='auc')]
    else:
        outputs = layers.Dense(num_classes, activation='softmax')(x)
        loss_fn = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy']

    model = models.Model(inputs, outputs)
    opt = optimizers.Adam(learning_rate=0.0001)
    model.compile(optimizer=opt, loss=loss_fn, metrics=metrics)
    return model

# ==========================================================
# 4ï¸âƒ£ Training Configuration
# ==========================================================
config = {
    'input_shape': (224, 224, 3),
    'batch_size': 16,  
    'rotation_range': 15,
    'zoom_range': 0.1,
    'horizontal_flip': True,
    'brightness_range': [0.8, 1.2]
}
# Add callbacks
callbacks = [
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    )
]
train_gen, val_gen, test_gen, num_classes = create_generators(train_df, val_df, test_df, config)

model = build_final_custom_cnn(input_shape=config['input_shape'], num_classes=num_classes)
# model.summary()

# ==========================================================
# 5ï¸âƒ£ Model Training
# ==========================================================
EPOCHS = 100

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks,  
    verbose=1
)

# ==========================================================
# 6ï¸âƒ£ Plot Training Curves
# ==========================================================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig(RESULTS_ROOT / "training_curves.png")
plt.show()

# ==========================================================
# 7ï¸âƒ£ Evaluation on Test Set
# ==========================================================
test_steps = len(test_gen)
test_preds = model.predict(test_gen, steps=test_steps, verbose=1)
true_labels = test_gen.classes
class_indices = test_gen.class_indices
class_names = list(class_indices.keys())

if num_classes == 2:
    y_pred_prob = test_preds.ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)
else:
    y_pred_prob = test_preds
    y_pred = np.argmax(test_preds, axis=1)

# Trim predictions if generator repeated last batch
if len(y_pred) > len(true_labels):
    y_pred = y_pred[:len(true_labels)]
    y_pred_prob = y_pred_prob[:len(true_labels)]

# ==========================================================
# 8ï¸âƒ£ Confusion Matrix & Classification Report
# ==========================================================
cm = confusion_matrix(true_labels, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig(RESULTS_ROOT / "confusion_matrix.png")
plt.show()

print("ðŸ“‹ Classification Report:")
print(classification_report(true_labels, y_pred, target_names=class_names))

# ==========================================================
# 9ï¸âƒ£ ROC Curve
# ==========================================================
plt.figure(figsize=(8, 6))
if num_classes == 2:
    fpr, tpr, _ = roc_curve(true_labels, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
else:
    y_true_bin = tf.keras.utils.to_categorical(true_labels, num_classes=num_classes)
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC={roc_auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig(RESULTS_ROOT / "roc_curve.png")
plt.show()
model.save("oral_cancer_model.h5")
```

```python
import os
from pathlib import Path
import itertools
import random
import json
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, roc_curve, auc

# ==========================================================
# 1ï¸âƒ£ Paths
# ==========================================================
DATASET_DIR = Path('/Volumes/Sarbajit/Model Optimization/Data/other data/Skin Leison/Augmented/Skin Leison')
RESULTS_ROOT = Path.cwd() / "CustomCNN_Skin"
RESULTS_ROOT.mkdir(parents=True, exist_ok=True)

# ==========================================================
# 2ï¸âƒ£ Build DataFrame & Split
# ==========================================================
def build_image_dataframe(dataset_dir: Path):
    filepaths = list(dataset_dir.glob("**/*.jpg")) + list(dataset_dir.glob("**/*.png")) + list(dataset_dir.glob("**/*.jpeg"))
    labels = [os.path.split(os.path.split(f)[0])[1] for f in filepaths]
    df = pd.DataFrame({'Filepath': [str(f) for f in filepaths], 'Label': labels})
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    return df

image_df = build_image_dataframe(DATASET_DIR)
print("Total images:", len(image_df))
print("Classes:", sorted(image_df['Label'].unique()))

# Stratified 70/15/15 split
train_val_df, test_df = train_test_split(image_df, train_size=0.85, random_state=42, stratify=image_df['Label'])
train_df, val_df = train_test_split(train_val_df, train_size=0.8235, random_state=42, stratify=train_val_df['Label'])
print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# ==========================================================
# 3ï¸âƒ£ Data Generators
# ==========================================================
def create_generators(train_df, val_df, test_df, config):
    num_classes = train_df['Label'].nunique()
    class_mode = 'binary' if num_classes == 2 else 'categorical'

    preprocess = tf.keras.applications.mobilenet_v2.preprocess_input

    train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        preprocessing_function=preprocess,
        rotation_range=config.get('rotation_range', 0),
        zoom_range=config.get('zoom_range', 0.0),
        horizontal_flip=config.get('horizontal_flip', False),
        brightness_range=config.get('brightness_range', None)
    )

    val_test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=preprocess)

    target_size = config['input_shape'][:2]
    batch_size = config.get('batch_size', 32)

    train_gen = train_datagen.flow_from_dataframe(
        train_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=True
    )

    val_gen = val_test_datagen.flow_from_dataframe(
        val_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    test_gen = val_test_datagen.flow_from_dataframe(
        test_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    return train_gen, val_gen, test_gen, num_classes

# ==========================================================
# ðŸ”¹ Attention Blocks (Keras-friendly)
# ==========================================================

def se_block(x, ratio=16):
    filters = x.shape[-1]
    se = layers.GlobalAveragePooling2D()(x)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, 1, filters))(se)
    return layers.Multiply()([x, se])

def cbam_block(x, ratio=8):
    channel = x.shape[-1]

    # Channel attention
    avg_pool = layers.GlobalAveragePooling2D()(x)
    max_pool = layers.GlobalMaxPooling2D()(x)
    shared_dense_one = layers.Dense(channel // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
    shared_dense_two = layers.Dense(channel, kernel_initializer='he_normal', use_bias=False)
    avg_out = shared_dense_two(shared_dense_one(avg_pool))
    max_out = shared_dense_two(shared_dense_one(max_pool))
    channel_attention = layers.Add()([avg_out, max_out])
    channel_attention = layers.Activation('sigmoid')(channel_attention)
    channel_attention = layers.Reshape((1, 1, channel))(channel_attention)
    x_ca = layers.Multiply()([x, channel_attention])

    # Spatial attention
    avg_pool_spatial = layers.Lambda(lambda t: tf.reduce_mean(t, axis=-1, keepdims=True))(x_ca)
    max_pool_spatial = layers.Lambda(lambda t: tf.reduce_max(t, axis=-1, keepdims=True))(x_ca)
    concat = layers.Concatenate(axis=-1)([avg_pool_spatial, max_pool_spatial])
    spatial_attention = layers.Conv2D(1, kernel_size=7, padding='same', activation='sigmoid')(concat)
    return layers.Multiply()([x_ca, spatial_attention])

def self_attention_block(x):
    """
    Safe self-attention using Lambda to wrap raw TF ops.
    Only use when spatial size is small (e.g., <= 14x14).
    """
    channels = x.shape[-1]
    f = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    g = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    h = layers.Conv2D(channels, 1, padding='same')(x)

    s = layers.Lambda(lambda tensors: tf.matmul(
        tf.reshape(tensors[0], [tf.shape(tensors[0])[0], -1, tf.shape(tensors[0])[-1]]),
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]]),
        transpose_b=True
    ))([f, g])

    beta = layers.Activation('softmax')(s)

    o = layers.Lambda(lambda tensors: tf.matmul(
        tensors[0],
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]])
    ))([beta, h])

    o = layers.Lambda(lambda tensors: tf.reshape(tensors[0], tf.shape(tensors[1])))([o, x])
    return layers.Add()([o, x])

# ==========================================================
# ðŸ”¹ Final Custom CNN Model â€” MEMORY-SAFE VERSION
# ==========================================================

def build_final_custom_cnn(input_shape=(224, 224, 3), num_classes=8):
    inputs = layers.Input(shape=input_shape)
    x = inputs

    filters = [32, 64, 128, 256]
    kernel_sizes = [3, 3, 3, 3]

    for i, (f, k) in enumerate(zip(filters, kernel_sizes)):
        x = layers.Conv2D(f, k, padding='same', kernel_initializer='he_normal',
                          kernel_regularizer=regularizers.l2(1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)

        # âœ… Apply SE in all blocks
        x = se_block(x, ratio=8)

        # âœ… Apply CBAM only in deeper blocks (i >= 2 â†’ 56x56 and smaller)
        if i >= 2:
            x = cbam_block(x, ratio=8)

        # âœ… Apply self-attention ONLY in the last block (14x14)
        if i == 3:
            x = self_attention_block(x)

        x = layers.MaxPooling2D(pool_size=2)(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)

    if num_classes == 2:
        outputs = layers.Dense(1, activation='sigmoid')(x)
        loss_fn = tf.keras.losses.BinaryCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy', tf.keras.metrics.AUC(name='auc')]
    else:
        outputs = layers.Dense(num_classes, activation='softmax')(x)
        loss_fn = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy']

    model = models.Model(inputs, outputs)
    opt = optimizers.Adam(learning_rate=0.0001)
    model.compile(optimizer=opt, loss=loss_fn, metrics=metrics)
    return model

# ==========================================================
# 4ï¸âƒ£ Training Configuration
# ==========================================================
config = {
    'input_shape': (224, 224, 3),
    'batch_size': 16,  
    'rotation_range': 15,
    'zoom_range': 0.1,
    'horizontal_flip': True,
    'brightness_range': [0.8, 1.2]
}
# Add callbacks
callbacks = [
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    )
]
train_gen, val_gen, test_gen, num_classes = create_generators(train_df, val_df, test_df, config)

model = build_final_custom_cnn(input_shape=config['input_shape'], num_classes=num_classes)
# model.summary()

# ==========================================================
# 5ï¸âƒ£ Model Training
# ==========================================================
EPOCHS = 100

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks,  
    verbose=1
)

# ==========================================================
# 6ï¸âƒ£ Plot Training Curves
# ==========================================================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig(RESULTS_ROOT / "training_curves.png")
plt.show()

# ==========================================================
# 7ï¸âƒ£ Evaluation on Test Set
# ==========================================================
test_steps = len(test_gen)
test_preds = model.predict(test_gen, steps=test_steps, verbose=1)
true_labels = test_gen.classes
class_indices = test_gen.class_indices
class_names = list(class_indices.keys())

if num_classes == 2:
    y_pred_prob = test_preds.ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)
else:
    y_pred_prob = test_preds
    y_pred = np.argmax(test_preds, axis=1)

# Trim predictions if generator repeated last batch
if len(y_pred) > len(true_labels):
    y_pred = y_pred[:len(true_labels)]
    y_pred_prob = y_pred_prob[:len(true_labels)]

# ==========================================================
# 8ï¸âƒ£ Confusion Matrix & Classification Report
# ==========================================================
cm = confusion_matrix(true_labels, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig(RESULTS_ROOT / "confusion_matrix.png")
plt.show()

print("ðŸ“‹ Classification Report:")
print(classification_report(true_labels, y_pred, target_names=class_names))

# ==========================================================
# 9ï¸âƒ£ ROC Curve
# ==========================================================
plt.figure(figsize=(8, 6))
if num_classes == 2:
    fpr, tpr, _ = roc_curve(true_labels, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
else:
    y_true_bin = tf.keras.utils.to_categorical(true_labels, num_classes=num_classes)
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC={roc_auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig(RESULTS_ROOT / "roc_curve.png")
plt.show()
model.save("skin_cancer_model.h5")
```

```python
import os
from pathlib import Path
import itertools
import random
import json
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, roc_curve, auc

# ==========================================================
# 1ï¸âƒ£ Paths
# ==========================================================
DATASET_DIR = Path('/Volumes/Sarbajit/Model Optimization/Data/Combined_Data')
RESULTS_ROOT = Path.cwd() / "CustomCNN_Combined"
RESULTS_ROOT.mkdir(parents=True, exist_ok=True)

# ==========================================================
# 2ï¸âƒ£ Build DataFrame & Split
# ==========================================================
def build_image_dataframe(dataset_dir: Path):
    filepaths = list(dataset_dir.glob("**/*.jpg")) + list(dataset_dir.glob("**/*.png")) + list(dataset_dir.glob("**/*.jpeg"))
    labels = [os.path.split(os.path.split(f)[0])[1] for f in filepaths]
    df = pd.DataFrame({'Filepath': [str(f) for f in filepaths], 'Label': labels})
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    return df

image_df = build_image_dataframe(DATASET_DIR)
print("Total images:", len(image_df))
print("Classes:", sorted(image_df['Label'].unique()))

# Stratified 70/15/15 split
train_val_df, test_df = train_test_split(image_df, train_size=0.85, random_state=42, stratify=image_df['Label'])
train_df, val_df = train_test_split(train_val_df, train_size=0.8235, random_state=42, stratify=train_val_df['Label'])
print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# ==========================================================
# 3ï¸âƒ£ Data Generators
# ==========================================================
def create_generators(train_df, val_df, test_df, config):
    num_classes = train_df['Label'].nunique()
    class_mode = 'binary' if num_classes == 2 else 'categorical'

    preprocess = tf.keras.applications.mobilenet_v2.preprocess_input

    train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        preprocessing_function=preprocess,
        rotation_range=config.get('rotation_range', 0),
        zoom_range=config.get('zoom_range', 0.0),
        horizontal_flip=config.get('horizontal_flip', False),
        brightness_range=config.get('brightness_range', None)
    )

    val_test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=preprocess)

    target_size = config['input_shape'][:2]
    batch_size = config.get('batch_size', 32)

    train_gen = train_datagen.flow_from_dataframe(
        train_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=True
    )

    val_gen = val_test_datagen.flow_from_dataframe(
        val_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    test_gen = val_test_datagen.flow_from_dataframe(
        test_df, x_col='Filepath', y_col='Label',
        target_size=target_size, class_mode=class_mode,
        batch_size=batch_size, shuffle=False
    )

    return train_gen, val_gen, test_gen, num_classes

# ==========================================================
# ðŸ”¹ Attention Blocks (Keras-friendly)
# ==========================================================

def se_block(x, ratio=16):
    filters = x.shape[-1]
    se = layers.GlobalAveragePooling2D()(x)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, 1, filters))(se)
    return layers.Multiply()([x, se])

def cbam_block(x, ratio=8):
    channel = x.shape[-1]

    # Channel attention
    avg_pool = layers.GlobalAveragePooling2D()(x)
    max_pool = layers.GlobalMaxPooling2D()(x)
    shared_dense_one = layers.Dense(channel // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)
    shared_dense_two = layers.Dense(channel, kernel_initializer='he_normal', use_bias=False)
    avg_out = shared_dense_two(shared_dense_one(avg_pool))
    max_out = shared_dense_two(shared_dense_one(max_pool))
    channel_attention = layers.Add()([avg_out, max_out])
    channel_attention = layers.Activation('sigmoid')(channel_attention)
    channel_attention = layers.Reshape((1, 1, channel))(channel_attention)
    x_ca = layers.Multiply()([x, channel_attention])

    # Spatial attention
    avg_pool_spatial = layers.Lambda(lambda t: tf.reduce_mean(t, axis=-1, keepdims=True))(x_ca)
    max_pool_spatial = layers.Lambda(lambda t: tf.reduce_max(t, axis=-1, keepdims=True))(x_ca)
    concat = layers.Concatenate(axis=-1)([avg_pool_spatial, max_pool_spatial])
    spatial_attention = layers.Conv2D(1, kernel_size=7, padding='same', activation='sigmoid')(concat)
    return layers.Multiply()([x_ca, spatial_attention])

def self_attention_block(x):
    """
    Safe self-attention using Lambda to wrap raw TF ops.
    Only use when spatial size is small (e.g., <= 14x14).
    """
    channels = x.shape[-1]
    f = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    g = layers.Conv2D(channels // 8 if channels >= 8 else 1, 1, padding='same')(x)
    h = layers.Conv2D(channels, 1, padding='same')(x)

    s = layers.Lambda(lambda tensors: tf.matmul(
        tf.reshape(tensors[0], [tf.shape(tensors[0])[0], -1, tf.shape(tensors[0])[-1]]),
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]]),
        transpose_b=True
    ))([f, g])

    beta = layers.Activation('softmax')(s)

    o = layers.Lambda(lambda tensors: tf.matmul(
        tensors[0],
        tf.reshape(tensors[1], [tf.shape(tensors[1])[0], -1, tf.shape(tensors[1])[-1]])
    ))([beta, h])

    o = layers.Lambda(lambda tensors: tf.reshape(tensors[0], tf.shape(tensors[1])))([o, x])
    return layers.Add()([o, x])

# ==========================================================
# ðŸ”¹ Final Custom CNN Model â€” MEMORY-SAFE VERSION
# ==========================================================

def build_final_custom_cnn(input_shape=(224, 224, 3), num_classes=33):
    inputs = layers.Input(shape=input_shape)
    x = inputs

    filters = [32, 64, 128, 256]
    kernel_sizes = [3, 3, 3, 3]

    for i, (f, k) in enumerate(zip(filters, kernel_sizes)):
        x = layers.Conv2D(f, k, padding='same', kernel_initializer='he_normal',
                          kernel_regularizer=regularizers.l2(1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)

        # âœ… Apply SE in all blocks
        x = se_block(x, ratio=8)

        # âœ… Apply CBAM only in deeper blocks (i >= 2 â†’ 56x56 and smaller)
        if i >= 2:
            x = cbam_block(x, ratio=8)

        # âœ… Apply self-attention ONLY in the last block (14x14)
        if i == 3:
            x = self_attention_block(x)

        x = layers.MaxPooling2D(pool_size=2)(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)

    if num_classes == 2:
        outputs = layers.Dense(1, activation='sigmoid')(x)
        loss_fn = tf.keras.losses.BinaryCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy', tf.keras.metrics.AUC(name='auc')]
    else:
        outputs = layers.Dense(num_classes, activation='softmax')(x)
        loss_fn = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1)
        metrics = ['accuracy']

    model = models.Model(inputs, outputs)
    opt = optimizers.Adam(learning_rate=0.0001)
    model.compile(optimizer=opt, loss=loss_fn, metrics=metrics)
    return model

# ==========================================================
# 4ï¸âƒ£ Training Configuration
# ==========================================================
config = {
    'input_shape': (224, 224, 3),
    'batch_size': 16,  
    'rotation_range': 15,
    'zoom_range': 0.1,
    'horizontal_flip': True,
    'brightness_range': [0.8, 1.2]
}
# Add callbacks
callbacks = [
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    )
]
train_gen, val_gen, test_gen, num_classes = create_generators(train_df, val_df, test_df, config)

model = build_final_custom_cnn(input_shape=config['input_shape'], num_classes=num_classes)
# model.summary()

# ==========================================================
# 5ï¸âƒ£ Model Training
# ==========================================================
EPOCHS = 100

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks,  
    verbose=1
)

# ==========================================================
# 6ï¸âƒ£ Plot Training Curves
# ==========================================================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig(RESULTS_ROOT / "training_curves.png")
plt.show()

# ==========================================================
# 7ï¸âƒ£ Evaluation on Test Set
# ==========================================================
test_steps = len(test_gen)
test_preds = model.predict(test_gen, steps=test_steps, verbose=1)
true_labels = test_gen.classes
class_indices = test_gen.class_indices
class_names = list(class_indices.keys())

if num_classes == 2:
    y_pred_prob = test_preds.ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)
else:
    y_pred_prob = test_preds
    y_pred = np.argmax(test_preds, axis=1)

# Trim predictions if generator repeated last batch
if len(y_pred) > len(true_labels):
    y_pred = y_pred[:len(true_labels)]
    y_pred_prob = y_pred_prob[:len(true_labels)]

# ==========================================================
# 8ï¸âƒ£ Confusion Matrix & Classification Report
# ==========================================================
cm = confusion_matrix(true_labels, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig(RESULTS_ROOT / "confusion_matrix.png")
plt.show()

print("ðŸ“‹ Classification Report:")
print(classification_report(true_labels, y_pred, target_names=class_names))

# ==========================================================
# 9ï¸âƒ£ ROC Curve
# ==========================================================
plt.figure(figsize=(8, 6))
if num_classes == 2:
    fpr, tpr, _ = roc_curve(true_labels, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
else:
    y_true_bin = tf.keras.utils.to_categorical(true_labels, num_classes=num_classes)
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC={roc_auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig(RESULTS_ROOT / "roc_curve.png")
plt.show()
model.save("combined_model.h5")
```
